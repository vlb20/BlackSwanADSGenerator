```python
import pandas as pd

df_accident = pd.read_csv("FARS/accident.csv", encoding="latin-1")
df_person = pd.read_csv("FARS/person.csv", encoding="latin-1")
df_vehicle = pd.read_csv("FARS/vehicle.csv", encoding="latin-1")

print(f"File ACCIDENT: {df_accident.shape[0]} righe, {df_accident.shape[1]} colonne")
print(f"File PERSON: {df_person.shape[0]} righe, {df_person.shape[1]} colonne")
print(f"File VEHICLE: {df_vehicle.shape[0]} righe, {df_vehicle.shape[1]} colonne")
```

```python
print("Colonne in ACCIDENT:", df_accident.columns.tolist())
print("\n")

print("Colonne in PERSON:", df_person.columns.tolist())
print("\n")

print("Colonne in VEHICLE:", df_vehicle.columns.tolist())
print("\n")
```

```python
accident_cols_to_keep = [
    'ST_CASE',
    'FATALS',
    'MONTHNAME',
    'DAY_WEEKNAME',
    'HOUR',
    'WEATHERNAME',
    'LGT_CONDNAME',
    'MAN_COLLNAME',
    'REL_ROADNAME',
    'TYP_INTNAME',
]

vehicle_cols_to_keep = [
    'ST_CASE',
    'VEH_NO',
    'BODY_TYPNAME',
    'MOD_YEAR',
    'DEFORMEDNAME',
    'P_CRASH1NAME',
    'P_CRASH2NAME',
    'P_CRASH3NAME',
    'SPEEDRELNAME',
    'V_CONFIGNAME',
    'J_KNIFENAME',
    'UNDEROVERRIDENAME',
    'CARGO_BTNAME',
    'HAZ_INVNAME',
    'PREV_DWINAME',
    'PREV_SPDNAME',
]

person_cols_to_keep = [
    'ST_CASE',
    'VEH_NO',
    'PER_NO',
    'PER_TYPNAME',
    'AGE',
    'SEXNAME',
    'DRINKINGNAME',
    'DRUGSNAME',
    'INJ_SEVNAME',
]

acc_df_filtered = df_accident[accident_cols_to_keep].copy()
veh_df_filtered = df_vehicle[vehicle_cols_to_keep].copy()
per_df_filtered = df_person[person_cols_to_keep].copy()

print("New columns shape ACCIDENT:", acc_df_filtered.shape[1])
print("New columns shape VEHICLE:", veh_df_filtered.shape[1])
print("New columns shape PERSON:", per_df_filtered.shape[1])
```

```python
df_merged = pd.merge(acc_df_filtered, veh_df_filtered, on='ST_CASE', how='left')

df_final = pd.merge(df_merged, per_df_filtered, on=['ST_CASE', 'VEH_NO'], how='left')

print("Dimensione del dataset finale:", df_final.shape)
print(df_final.head())
```

```python
print(df_final.isnull().sum())
```

```python
cols_to_remove = ['PREV_DWINAME', 'PREV_SPDNAME']
df_final.drop(columns=cols_to_remove, inplace=True)

print(f"Rimosse le colonne: {cols_to_remove}")
```

```python
print(df_final.isnull().sum())
```

```python
df_final.dropna(subset=['PER_NO'], inplace=True)
```
